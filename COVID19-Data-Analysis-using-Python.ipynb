#### Step 6: Find the first derivative of the curve.  
First derivative (instantaneous rate of change) of a function tells the slope of the tangent of the curve.  
Tangent +ve : increase in curve    
Tangent -ve : decrease in curve  

```python
corona_aggregated.loc['China'].diff().plot()
```

The graph shows us the infection rate day by day and we want the maximum number.

```python
corona_aggregated.loc['China'].diff().max()
#in only 24hrs, difference was 15136.
```

```python
corona_aggregated.loc['India'].diff().max()
#in only 24hrs, difference was 1893.
```

```python
corona_aggregated.loc['Italy'].diff().max()
#in only 24hrs, difference was 6557.
```

#### Step 7: Find the maximum infection rate for all the countries.


```python
countries = list(corona_aggregated.index)
max_infections = []
for c in countries:
    max_infections.append(corona_aggregated.loc[c].diff().max())
max_infections
```

```python
corona_aggregated['max_infections'] = max_infections
corona_aggregated.head(5)
```

#### Step 8: Prepare dataframe with countries and infections

```python
corona_data=pd.DataFrame(corona_aggregated['max_infections'])
corona_data
```

### Let's compare GDP and maxi_infections

```python
x=data['GDP per capita']
y=data['max_infections']
sns.scatterplot(x,y)

#we can see values need different scaling
```

```python
# We'll apply log scaling to y
x=data['GDP per capita']
y=data['max_infections']
sns.scatterplot(x,np.log(y))

#we can see +ve correlation
```

```python
#RegPlot
x=data['GDP per capita']
y=data['max_infections']
sns.regplot(x,np.log(y))

#line fitter, +ve slope seen
```
